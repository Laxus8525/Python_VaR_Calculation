```python
import pandas as pd
import numpy as np
from scipy.stats import norm
```

### Data

```python
df = pd.read_excel('Data_VaR.xlsm',sheet_name = 'Data')
df = df.set_index('Date')
df
```

### Price Change

### stock price

```python
stock = df[['MSFT','FORD','BOA','AAPL']]
stock = stock.apply(lambda x: round(x,2))
stock
```

```python
stock_dac = stock.pct_change().dropna()
stock_dac
```

### SOFR

```python
sofr = df.loc[:, '1D':'40Y']
sofr
```

```python
# absolute return
sofr_dac = sofr.diff().dropna()
sofr_dac
```

### LIBOR

```python
libor = df.loc[:, '1D.1':'40Y.1']
libor
```

```python
# absolute return
libor_dac = libor.diff().dropna()
libor_dac
```

```python
# concat table
dac = stock_dac.join(sofr_dac).join(libor_dac)
mean = dac.describe().iloc[1]
std = dac.describe().iloc[2]
```

```python
dac
```

```python
dac.describe()
```

```python
mean
```

```python
std
```

```python
# subtract mean 
submean_dac = (dac - dac.mean())
submean_dac
```

```python
# standardized
standard_dac = (dac - dac.mean())/dac.std()
standard_dac
```

### Matrix

```python
# varaince matrix
V = submean_dac.cov()
V
```

```python
# Correlation Matrix
C = standard_dac.corr()
C
```

```python
# Cholesky Decomposition
decomposition = pd.DataFrame(np.linalg.cholesky(C))
decomposition
```

### SwapPricer

```python
data = pd.read_excel('Data_VaR.xlsm',sheet_name = 'SwapPV01')
data
```

```python
# sofr curve
sofr_curve = data[['Unnamed: 3','Unnamed: 4']].dropna()
sofr_curve.columns = ['SOFR_Curve', 'pv01']
sofr_curve = sofr_curve[1:]
sofr_curve = sofr_curve.set_index('SOFR_Curve')
sofr_curve
```

```python
# libor curve
libor_curve = data[['Unnamed: 7','Unnamed: 8']].dropna()
libor_curve.columns = ['LIBOR_Curve', 'pv01']
libor_curve = libor_curve[1:]
libor_curve = libor_curve.set_index('LIBOR_Curve')
libor_curve
```

```python
# total
sofr_total = sofr_curve.sum().values
libor_total = libor_curve.sum().values
print('sofr_total:{},libor_total:{}'.format(sofr_total,libor_total))
```

### linear weight

```python
df1 = pd.DataFrame(index=stock.columns)
df1['Weight'] = 1000000
df2 = pd.DataFrame(index=sofr_curve.index)
df2['Weight'] = sofr_curve / 0.0001
df3 = pd.DataFrame(index=libor_curve.index)
df3['Weight'] = libor_curve / 0.0001
linear_weight  = pd.concat([df1,df2,df3])
linear_weight
```

### Parametric-VaR

```python
alpha = 5/100 # 95%
expected_PNL = np.matmul(mean.values,linear_weight.Weight)
std_PNL = np.sqrt(np.matmul(np.matmul(linear_weight.Weight,V.values),linear_weight.Weight))
Param_VaR = abs(expected_PNL + norm.ppf(alpha) * std_PNL)
```

```python
print("Expected Value of PNL: {}".format(expected_PNL))
print("Stdev of PNL: {}".format(std_PNL))
print("Parametric Var: {}".format(Param_VaR))
```

### MonteCarlo-VaR

```python
# uniform distribution random variable
random = pd.DataFrame(np.random.uniform(size = (10000, 64)),index=range(1,10001), columns= range(1,65))
random 
```

```python
# norm.ppf(random) cannot work something wrong with the first columns
l = []
for i in random.iloc[:,0]:
    l.append(norm.ppf(i))
l = np.array(l)
```

```python
# uniform random -> std norm random by inverse norm
norm_random = pd.DataFrame(l,columns=[0])
norm_random.loc[:,range(1,64)] = norm.ppf(random.iloc[:,1:])
norm_random
```

```python
# correlated std normal variates
corr_random = pd.DataFrame(np.matmul(decomposition.values,norm_random.T.values).T)
corr_random.columns = linear_weight.index
corr_random
```

```python
mean.index =  linear_weight.index
std.index = linear_weight.index
```

```python
risk_factor = mean + std * corr_random
risk_factor
```

```python
PnL = pd.DataFrame(np.matmul(risk_factor.values,linear_weight.Weight),columns=['PnL'])
PnL = PnL.sort_values(by='PnL',ascending=True)
PnL = PnL.reset_index(drop=True)
PnL
```

```python
MC_VaR = float(abs(PnL.iloc[499].values))
MC_VaR
```
